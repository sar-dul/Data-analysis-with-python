## 7. Conduct hypothesis testing on target features using chi-square tests and test for normality using Shapiro test or Kolmogorov-Smirnov test.

```python
import scipy.stats as stats
```

```python
# Function for chi-square test

def chi_square_test(column):
    print("Chi-square Test between SEVERITYCODE and " + column)
    severity_df = pd.crosstab(df['SEVERITYCODE'], df[column])
    chi2_stat, p_val, dof, expected = stats.chi2_contingency(severity_df)
    print("DOF:", dof)
    print("p-value:", p_val)
    
    if p_val < 0.05:
        print("There is strong relationship")
    else: 
        print("There is no relationship")
```

```python
# All object datatype columns

df.select_dtypes(include=['object']).columns
```

```python
# Applying chi square test for categorical columns

categorical_columns = ['ADDRTYPE', 'COLLISIONTYPE', 'JUNCTIONTYPE', 
                       'UNDERINFL', 'WEATHER', 'ROADCOND', 'LIGHTCOND', 
                       'ST_COLDESC', 'HITPARKEDCAR']
for column in categorical_columns:
    chi_square_test(column)
    print("-" * 70)
```

```python
# Function of Shapiro-Wilk test for normality

def shapiro_wilk_test(column):
    print("Shapiro-Wilk test for " + column)
    shapiro_stat, shapiro_p_val = stats.shapiro(df[column])
    print("Shapiro-Wilk statistic:", shapiro_stat)
    print("p-value:", shapiro_p_val)
    
    if shapiro_p_val > 0.05:
        print("Normally distributed.")
    else:
        print("Not normally distributed")
```

```python
# All numerical columns

df.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Applying shapiro wilk test for numerical columns

numerical_columns = [ 'PERSONCOUNT', 'PEDCOUNT', 'PEDCYLCOUNT', 
                     'VEHCOUNT', 'INJURIES', 'SERIOUSINJURIES', 'FATALITIES', 
                     'SDOT_COLCODE', 'STCOLCODE', 'SEGLANEKEY', 'CROSSWALKKEY']
for column in numerical_columns:
    shapiro_wilk_test(column)
    print("-" * 70)
```

```python
# Function of Kolmogorov-Smirnov test for normality

def kolmogorov_smirnov_test(column):
    print("Kolmogorov-Smirnov test for " + column)
    ks_stat, ks_p_val = stats.kstest(df[column], 'norm')
    print("Kolmogorov-Smirnov statistic:", ks_stat)
    print("p-value:", ks_p_val)
    
    if ks_p_val > 0.05:
        print("Normally distributed.")
    else:
        print("Not normally distributed")
```

```python
# Applying shapiro wilk test for numerical columns

for column in numerical_columns:
    kolmogorov_smirnov_test(column)
    print("-" * 70)
```

## 8. Conduct correlation testing, using Pearson correlation test for linear relationships between features and determining the appropriate test for non-linear relationships (e.g., through visual inspection with scatterplots).

```python
# Function of Pearson correlation test

def pearson_correlation_test(column1, column2):
    print("Correlation between " + column1 + " and "+ column2)
    corr_coef, p_value = stats.pearsonr(df[column1], df[column2])
    print("KP's Correlation Coefficient:", corr_coef)
    print("p-value:", p_value)
    
    if ((corr_coef > 0.05) or (corr_coef < -0.05)):
        print("Higher Correlation")
    elif ((corr_coef > 0.3) or (corr_coef < -0.3)):
        print("Moderate Correlation")
    else:
        print("Low Correlation")
        return "low"
```

```python
low_correlation_columns = [] # dictionary of low correlation columns

# Applying Pearson correlation test for numerical columns

for column1 in numerical_columns:
    for column2 in numerical_columns:
        corr = pearson_correlation_test(column1, column2)
        if corr == "low":
            low_correlation_columns.append([column1, column2])
        print("-" * 70)
```

```python
# Scatter plots of non correlated columns (visual inspection)

def scatter_plot(column1, column2):
    sns.scatterplot(x=column1, y=column2, data=df)
    plt.title('Scatter Plot between ' + column1 + ' and ' + column2)
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.show()
```

```python
#  Visualizing low correlated data in scatter plot

for each_rel in low_correlation_columns:
    scatter_plot(each_rel[0], each_rel[1])
    
```
